```python
data=pd.read_csv('countries.csv')
ch = data[data.country == 'China']
us = data[data.country == 'United States']
plt.plot(us.year,us.population/ us.population.iloc[0] *100,'r--')
plt.plot(ch.year,ch.population/ ch.population.iloc[0]*100,'g-o')
plt.xlabel('Year')
plt.ylabel('Population growth first year = 100')
plt.legend(['This is america','This is china'])
plt.title('UsA and Chnia growth in %')
```
